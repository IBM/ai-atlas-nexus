# Jurisdiction Slot on Concepts

This notebook demonstrates the use of the `jurisdiction` slot on `Concept` entities in the AI Atlas Nexus ontology. The `jurisdiction` slot enables specifying which ISO 3166-1 countries a concept applies to, using a dynamic enum sourced from the DPV Location ontology.

**Key features:**
- Multi-valued: a concept can apply to multiple jurisdictions
- Dynamic enum: values are sourced from the DPV Location ontology (not hardcoded)
- Semantic linking: slot URI `dpv:hasJurisdiction` integrates with semantic web standards

## Setup

```python
import json
import sys

from ai_atlas_nexus.ai_risk_ontology.datamodel.ai_risk_ontology import (
    Concept,
    Jurisdiction,
    Risk,
    Principle,
    Term,
)

print("✓ Imported Concept classes and Jurisdiction enum")
```

## Example 1: EU AI Act Risk with Geographic Scope

Create a `Risk` concept that applies specifically to EU jurisdictions due to the EU AI Act regulatory scope.

```python
# Example 1: EU-scoped risk under AI Act

eu_transparency_risk = Risk(
    id="eu-ai-act-risk-transparency-violation",
    name="EU AI Act: Transparency Requirements Violation",
    description="Risk that high-risk AI systems fail to provide required transparency to users about AI involvement, capabilities, and limitations, violating EU AI Act Article 13.",
    hasJurisdiction=[
        Jurisdiction("AT"),  # Austria
        Jurisdiction("BE"),  # Belgium
        Jurisdiction("BG"),  # Bulgaria
        Jurisdiction("HR"),  # Croatia
        Jurisdiction("CY"),  # Cyprus
        Jurisdiction("CZ"),  # Czech Republic
        Jurisdiction("DK"),  # Denmark
        Jurisdiction("EE"),  # Estonia
        Jurisdiction("FI"),  # Finland
        Jurisdiction("FR"),  # France
        Jurisdiction("DE"),  # Germany
        Jurisdiction("GR"),  # Greece
        Jurisdiction("HU"),  # Hungary
        Jurisdiction("IE"),  # Ireland
        Jurisdiction("IT"),  # Italy
        Jurisdiction("LV"),  # Latvia
        Jurisdiction("LT"),  # Lithuania
        Jurisdiction("LU"),  # Luxembourg
        Jurisdiction("MT"),  # Malta
        Jurisdiction("NL"),  # Netherlands
        Jurisdiction("PL"),  # Poland
        Jurisdiction("PT"),  # Portugal
        Jurisdiction("RO"),  # Romania
        Jurisdiction("SK"),  # Slovakia
        Jurisdiction("SI"),  # Slovenia
        Jurisdiction("ES"),  # Spain
        Jurisdiction("SE"),  # Sweden
    ],
)

print(f"⚠️  Created risk: {eu_transparency_risk.name}")
print(f"   ID: {eu_transparency_risk.id}")
print(f"   Applies to {len(eu_transparency_risk.hasJurisdiction)} EU member states")
print(f"\n   Jurisdictions (ISO 3166-1 alpha-2):")
print(f"   {', '.join(sorted([str(j) for j in eu_transparency_risk.hasJurisdiction]))}")
```

```python
# Serialize the risk to JSON to show the schema structure
risk_json = eu_transparency_risk.model_dump(mode='json', exclude_none=True)
print("📄 JSON representation:")
print(json.dumps(risk_json, indent=2))
```

## Example 2: Global Capability Concept with Multi-Region Applicability

Create a `CapabilityConcept` that applies to multiple jurisdictions with different regulatory regimes (EU, USA, and Asia-Pacific).

```python
# Example 2: Global AI capability spanning multiple regulatory regimes
from ai_atlas_nexus.ai_risk_ontology.datamodel.ai_risk_ontology import CapabilityConcept

data_residency_capability = CapabilityConcept(
    id="capability-data-residency-compliance",
    name="Data Residency Compliance Capability",
    description="Capability of AI systems to comply with data residency requirements across different jurisdictions. "
                "This affects GDPR (EU), CCPA (US), PDPA (Singapore), and PIPEDA (Canada) compliance.",
    hasJurisdiction=[
        # EU regulatory zone
        Jurisdiction("DE"),  # Germany (GDPR)
        Jurisdiction("FR"),  # France (GDPR)
        Jurisdiction("IE"),  # Ireland (GDPR)
        Jurisdiction("NL"),  # Netherlands (GDPR)
        # North American regulatory zone
        Jurisdiction("US"),  # USA (CCPA, state laws)
        Jurisdiction("CA"),  # Canada (PIPEDA)
        # Asia-Pacific regulatory zone
        Jurisdiction("JP"),  # Japan (APPI)
        Jurisdiction("SG"),  # Singapore (PDPA)
        Jurisdiction("AU"),  # Australia (Privacy Act)
        Jurisdiction("NZ"),  # New Zealand (Privacy Act)
    ],
)

print(f"✨ Created capability: {data_residency_capability.name}")
print(f"   ID: {data_residency_capability.id}")
print(f"   Applies to {len(data_residency_capability.hasJurisdiction)} jurisdictions")
print(f"\n   Jurisdictions (ISO 3166-1 alpha-2):")

# Group by region for readability
regions = {
    "🇪🇺 EU": ["DE", "FR", "IE", "NL"],
    "🇺🇸 North America": ["US", "CA"],
    "🌏 Asia-Pacific": ["JP", "SG", "AU", "NZ"],
}

jurisdictions_set = set(str(j) for j in data_residency_capability.hasJurisdiction)
for region, codes in regions.items():
    included = [c for c in codes if c in jurisdictions_set]
    if included:
        print(f"   {region}: {', '.join(included)}")
```

```python
# Show the capability in JSON format
capability_json = data_residency_capability.model_dump(mode='json', exclude_none=True)
print("📄 JSON representation:")
print(json.dumps(capability_json, indent=2))
```

## Comparison: Filtering Concepts by Jurisdiction

Demonstrate how the jurisdiction slot enables filtering concepts by region.

```python
# Create a small collection of concepts with different jurisdictions
concepts = [
    eu_transparency_risk,
    data_residency_capability,
    CapabilityConcept(
        id="capability-us-ai-governance",
        name="US AI Governance Compliance",
        description="AI system capability to comply with US-specific AI governance frameworks and executive orders.",
        hasJurisdiction=[Jurisdiction("US")],
    ),
]

print("📚 Concepts collection:")
for concept in concepts:
    print(f"  ✓ {concept.name}")

# Find all concepts applicable in Germany
print(f"\n🔍 Concepts applicable in Germany (DE):")
de_concepts = [
    c for c in concepts 
    if c.hasJurisdiction and any(str(j) == "DE" for j in c.hasJurisdiction)
]
for concept in de_concepts:
    print(f"  ✓ {concept.name}")
    print(f"    Applicable jurisdictions: {len(concept.hasJurisdiction)}")

# Find all concepts applicable in USA
print(f"\n🔍 Concepts applicable in USA (US):")
us_concepts = [
    c for c in concepts 
    if c.hasJurisdiction and any(str(j) == "US" for j in c.hasJurisdiction)
]
for concept in us_concepts:
    print(f"  ✓ {concept.name}")
    print(f"    Applicable jurisdictions: {len(concept.hasJurisdiction)}")
```

## Schema Integration

The `jurisdiction` slot is defined as:
- **Range**: `Jurisdiction` (dynamic enum from DPV Location ontology)
- **Multi-valued**: `true` (a concept can apply to multiple jurisdictions)
- **Slot URI**: `dpv:hasJurisdiction` (semantic web alignment)
- **Applies to**: All `Concept` subtypes (Principle, Risk, Term, etc.)

The `Jurisdiction` enum is sourced from:
- **Ontology**: DPV Location (https://w3id.org/dpv/loc)
- **Values**: ISO 3166-1 countries (all subclasses of `dpv:Country`)
- **Format**: Two-letter country codes (e.g., "US", "DE", "FR")
